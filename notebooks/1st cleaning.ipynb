```python
import matplotlib.pyplot as plt
import pandas as pd
plt.figure(figsize=(12,10))
plt.hist(df['Price Per Unit'],bins=7)
plt.xlabel('price per units')
plt.ylabel('total value counts of products')
plt.show()
```

```python
#df.groupby('Location')['Total Spent'].sum()
import matplotlib.pyplot as plt
spent_by_location=df.groupby('Location')['Total Spent'].sum()
index=spent_by_location.index
values=spent_by_location.values
def abs(pct):
    total=sum(values)
    absolute=int(round(pct*total/100))
    return f'{absolute}'
plt.pie(values,  labels=index,autopct=abs)
plt.show()
```
